```python
captioner1 = pipeline(task="image-to-text",
                     model="microsoft/git-base-coco")

captioner2 = pipeline(task="image-to-text",
                     model="nlpconnect/vit-gpt2-image-captioning")

captioner3 = pipeline(task="image-to-text",
                     model="Salesforce/blip-image-captioning-base")
```
